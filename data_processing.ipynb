```
# 코딩수정할때마다 맨 처음부터 실행 안 하려면
import pandas as pd

df = pd.read_csv('SME_Processed.csv')
```

```
# txt는 복잡해서 csv로 바꾸려고함
# test = pd.read_csv('testfile.txt', sep=" ", delimiter = '\t') 이런 방법도 있다

def txt_to_csv(txt,csv): #문제 있음 수정해야함
  df = pd.read_fwf(txt, header=None) # 첫 행이 열이름으로 지정되는문제. txt파일 불러오는 것 부터 문제였다. 여기서 헤더를 None으로 줬어야 해결이 가능했던 것이다!
  df.to_csv(csv, header=False,index=False)
  return df
```

```
# 잘 작동되는지 testfile.txt로 살펴보기

t = txt_to_csv('testfile.txt','test.csv')
t # txt파일이 csv로 성공적으로 변환되였고 column이름은 0,1,2... 열의 순서로 지정되었다.
```
